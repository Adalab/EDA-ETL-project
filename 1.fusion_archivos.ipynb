```python
def sumar_columnas (*args):
     
     set_tupla = list(set(args))
     set_tupla.remove(np.nan)
     
     
     if len(set_tupla) == 0:
          return np.nan
     elif len(set_tupla) > 1:
          return ", ".join(set_tupla)

     else:
          return set_tupla[0]
```
